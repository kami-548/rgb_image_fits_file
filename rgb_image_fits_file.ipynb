```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
# astropy visulization
from astropy.visualization import make_lupton_rgb
```

```python
# open fits files
r = fits.open('./arp_r.fits')
i = fits.open('./arp_i.fits')
g = fits.open('./arp_g.fits')
```

```python
# read image data
r_data = r[0].data
i_data = i[0].data
g_data = g[0].data
# 0 is for primary HDU
```

```python
#for plotting
plt.imshow(np.log(r_data))
plt.show()
```

```python
# for RGB file
rgb = make_lupton_rgb(r_data, i_data, g_data)
plt.imshow(rgb)
plt.show()
```

```python
# make image more clear
rgb = make_lupton_rgb(r_data, i_data, g_data, Q = 10, stretch= 0.5)
plt.imshow(rgb)
plt.show()
```
